## 2. ¿Qué es Python?

Python es un lenguaje de programación **interpretado**, **de alto nivel** y
**de propósito general** creado por Guido van Rossum en 1991.

Su filosofía se resume en el documento llamado *"El Zen de Python"*,
que está integrado directamente en el lenguaje:


````python
# El Zen de Python — la filosofía del lenguaje
import this
````

### ¿Por qué Python y no R o Excel?

No es que Python sea "mejor" — cada herramienta tiene su lugar.
La ventaja de Python es que **integra todo en un solo flujo de trabajo**:


````python
# Demostración rápida: en R necesitarías cambiar de herramienta,
# en Python todo vive en el mismo script.

# Paso 1: Cargar datos (como Excel)
datos = {
    "poliza": ["P-001", "P-002", "P-003", "P-004", "P-005"],
    "prima_anual": [2400, 5600, 1900, 12000, 3200],
    "edad_asegurado": [28, 45, 32, 62, 38]
}
df = pd.DataFrame(datos)

# Paso 2: Analizar (como R)
print("Estadísticas de primas:")
print(df["prima_anual"].describe().round(0))

# Paso 3: Visualizar (como ggplot)
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(df["poliza"], df["prima_anual"], color="#0D7A8A", edgecolor="white")
ax.set_title("Prima anual por póliza", fontsize=13, fontweight="bold")
ax.set_ylabel("Prima (MXN)")
ax.set_xlabel("Póliza")
plt.tight_layout()
plt.show()

print("✓ En Python: cargamos, analizamos y graficamos sin salir del notebook.")
````

### Tipos básicos de Python


````python
# ── int: Números enteros ────────────────────────────────────────────────────
edad = 35
num_siniestros = 2
year = 2026

print(type(edad))         # <class 'int'>
print(f"Edad: {edad}, Siniestros: {num_siniestros}")
````

````python
# ── float: Números decimales ────────────────────────────────────────────────
prima = 3_450.75
tasa_mortalidad = 0.00312
factor_ajuste   = 1.085

print(type(prima))        # <class 'float'>
print(f"Prima: ${prima:,.2f}")
print(f"Tasa de mortalidad: {tasa_mortalidad:.4%}")   # Formato porcentaje
````

````python
# ── str: Cadenas de texto ───────────────────────────────────────────────────
nombre_asegurado = "María García Hernández"
ramo = 'Gastos Médicos Mayores'    # Comillas simples o dobles, ambas funcionan
numero_poliza = "GMM-2026-00145"

print(type(nombre_asegurado))      # <class 'str'>
print(f"Póliza {numero_poliza} — Titular: {nombre_asegurado}")
print(f"Ramo: {ramo}")
print(f"Longitud del nombre: {len(nombre_asegurado)} caracteres")
````

````python
# ── bool: Valores lógicos ───────────────────────────────────────────────────
poliza_activa   = True
tiene_deducible = False
es_mayor_de_edad = 35 > 18    # Python evalúa la expresión y guarda el resultado

print(type(poliza_activa))    # <class 'bool'>
print(f"¿Póliza activa?: {poliza_activa}")
print(f"¿Tiene deducible?: {tiene_deducible}")
print(f"¿Mayor de edad?: {es_mayor_de_edad}")
````

````python
# ── NoneType: ausencia de valor ─────────────────────────────────────────────
# Equivalente a NULL en R o SQL, NA en R
beneficiario = None     # La póliza aún no tiene beneficiario registrado

print(type(beneficiario))         # <class 'NoneType'>
print(f"Beneficiario: {beneficiario}")
print(f"¿Tiene beneficiario?: {beneficiario is not None}")
````

````python
# ── Conversión entre tipos (casting) ────────────────────────────────────────
edad_str = "42"            # Llega como texto desde un CSV
edad_int = int(edad_str)   # Convertimos a entero para operar con ella
prima_str = "3450.75"
prima_float = float(prima_str)

print(f"Edad como texto: '{edad_str}' (tipo: {type(edad_str).__name__})")
print(f"Edad como entero: {edad_int} (tipo: {type(edad_int).__name__})")
print(f"Prima como float: {prima_float:,.2f}")

# Error común: intentar operar texto con número
# edad_int + edad_str   # ← TypeError: esto daría error
# La solución: convertir primero con int() o float()
````

## 5. Operaciones Básicas con Contexto Actuarial

Veamos cómo Python resuelve cálculos típicos que un actuario hace a diario.


````python
# ── Cálculo de prima con recargo de seguridad ───────────────────────────────
suma_asegurada  = 1_000_000   # 1 millón MXN
tasa_riesgo     = 0.018       # 1.8% tasa técnica
recargo_seg     = 0.12        # 12% de recargo de seguridad
factor_iva      = 1.16        # IVA 16%

prima_pura       = suma_asegurada * tasa_riesgo
prima_comercial  = prima_pura * (1 + recargo_seg)
prima_con_iva    = prima_comercial * factor_iva

print("─" * 42)
print("  CÁLCULO DE PRIMA — Seguro de Vida")
print("─" * 42)
print(f"  Suma asegurada    : ${suma_asegurada:>12,.2f}")
print(f"  Prima pura        : ${prima_pura:>12,.2f}")
print(f"  Recargo seguridad : {recargo_seg:.0%}")
print(f"  Prima comercial   : ${prima_comercial:>12,.2f}")
print(f"  Prima con IVA     : ${prima_con_iva:>12,.2f}")
print("─" * 42)
````

````python
# ── Cálculo de frecuencia y severidad ───────────────────────────────────────
# Métricas fundamentales en seguros de daños
total_polizas   = 8_500
total_siniestros = 374
monto_total_siniestros = 12_450_000  # MXN

frecuencia = total_siniestros / total_polizas
severidad  = monto_total_siniestros / total_siniestros
prima_pura_calculada = frecuencia * severidad

print("─" * 46)
print("  INDICADORES DE CARTERA — Autos 2025")
print("─" * 46)
print(f"  Pólizas expuestas   : {total_polizas:>8,}")
print(f"  Total siniestros    : {total_siniestros:>8,}")
print(f"  Frecuencia          : {frecuencia:>8.4f}  ({frecuencia:.2%})")
print(f"  Severidad promedio  : ${severidad:>10,.0f}")
print(f"  Prima pura implícita: ${prima_pura_calculada:>10,.2f}")
print("─" * 46)
````

````python
# ── Operadores de comparación y lógicos ─────────────────────────────────────
# Son la base de cualquier regla de underwriting

edad = 67
suma = 2_500_000
tiene_preexistencias = True

# Operadores de comparación: ==  !=  >  <  >=  <=
es_mayor_65 = edad > 65
supera_limite = suma > 2_000_000

# Operadores lógicos: and  or  not
requiere_revision = es_mayor_65 and supera_limite
acepta_automatico = not requiere_revision or not tiene_preexistencias

print(f"Edad: {edad}, ¿mayor de 65?: {es_mayor_65}")
print(f"Suma: ${suma:,}, ¿supera límite?: {supera_limite}")
print(f"¿Requiere revisión manual?: {requiere_revision}")
print(f"¿Se acepta automáticamente?: {acepta_automatico}")
````

````python
# ── f-strings: formato de texto ─────────────────────────────────────────────
# Las f-strings son la forma moderna de formatear texto en Python (Python 3.6+)
# Son esenciales para generar reportes y mensajes legibles

nombre = "Carlos Mendoza"
prima  = 4_325.50
fecha  = "2026-04-15"

# Formato básico
print(f"Titular: {nombre}")

# Formato numérico: ,:  separador de miles   .2f: 2 decimales
print(f"Prima anual: ${prima:,.2f}")

# Alineación: > alinea a la derecha, < a la izquierda
print(f"{'Campo':<20} {'Valor':>15}")
print(f"{'Nombre':<20} {nombre:>15}")
print(f"{'Prima anual':<20} ${prima:>14,.2f}")
print(f"{'Fecha emisión':<20} {fecha:>15}")
````

## 6. Nuestro Primer DataFrame Actuarial

Un **DataFrame** de Pandas es una tabla bidimensional — piénsalo como
una hoja de Excel viva en Python. Es la estructura de datos más importante
que usaremos durante todo el diplomado.

Aquí trabajamos con una mini-cartera de Gastos Médicos Mayores.


````python
# ── Crear un DataFrame desde un diccionario ─────────────────────────────────
import pandas as pd

cartera_gmm = {
    "id_poliza":       ["GMM-001", "GMM-002", "GMM-003", "GMM-004",
                        "GMM-005", "GMM-006", "GMM-007", "GMM-008"],
    "nombre":          ["Ana Torres", "Luis Pérez", "María García",
                        "Carlos Ruiz", "Rosa Méndez", "Jorge Silva",
                        "Elena Castro", "Pedro Vega"],
    "edad":            [28, 45, 32, 62, 38, 51, 29, 55],
    "sexo":            ["F", "M", "F", "M", "F", "M", "F", "M"],
    "suma_asegurada":  [500_000, 1_000_000, 300_000, 2_000_000,
                        500_000, 1_500_000, 300_000, 1_000_000],
    "prima_anual":     [2_400, 6_200, 1_900, 14_500,
                        3_100, 8_800, 1_850, 7_300],
    "siniestros_2025": [0, 1, 0, 3, 0, 1, 0, 2],
    "monto_siniestros":[0, 45_000, 0, 380_000, 0, 92_000, 0, 175_000],
    "activa":          [True, True, True, True, False, True, True, False],
}

df = pd.DataFrame(cartera_gmm)
print(f"Cartera cargada: {len(df)} pólizas, {len(df.columns)} columnas")
print()
df
````

````python
# ── Exploración inicial — lo primero que hacemos con cualquier dataset ───────

print("── FORMA DEL DATAFRAME ───────────────────────────────")
print(f"  Filas: {df.shape[0]}  |  Columnas: {df.shape[1]}")

print("── TIPOS DE DATOS ────────────────────────────────────")
print(df.dtypes)

print("── PRIMEROS 3 REGISTROS (df.head(3)) ─────────────────")
print(df.head(3).to_string())

print("── ÚLTIMOS 3 REGISTROS (df.tail(3)) ──────────────────")
print(df.tail(3).to_string())
````

````python
# ── Estadísticas descriptivas automáticas ───────────────────────────────────
print("Estadísticas de columnas numéricas:")
print()
df.describe().round(2)
````

````python
# ── Acceder a columnas individuales ─────────────────────────────────────────
# Sintaxis: df["nombre_columna"]  — devuelve una Serie de Pandas

primas = df["prima_anual"]
print("Serie de primas:")
print(primas)
print(f"Tipo: {type(primas)}")
````

````python
# ── Cálculos sobre columnas ─────────────────────────────────────────────────
print("─" * 48)
print("  RESUMEN DE LA CARTERA GMM")
print("─" * 48)
print(f"  Total pólizas         : {len(df)}")
print(f"  Pólizas activas       : {df['activa'].sum()}")
print(f"  Prima total anual     : ${df['prima_anual'].sum():>12,.0f}")
print(f"  Prima promedio        : ${df['prima_anual'].mean():>12,.2f}")
print(f"  Prima máxima          : ${df['prima_anual'].max():>12,.0f}")
print(f"  Prima mínima          : ${df['prima_anual'].min():>12,.0f}")
print(f"  Suma asegurada total  : ${df['suma_asegurada'].sum():>12,.0f}")
print(f"  Total siniestros      : {df['siniestros_2025'].sum()}")
print(f"  Monto total siniestros: ${df['monto_siniestros'].sum():>12,.0f}")
print("─" * 48)

# Siniestralidad (loss ratio)
primas_total = df['prima_anual'].sum()
siniestros_total = df['monto_siniestros'].sum()
loss_ratio = siniestros_total / primas_total
print(f"Siniestralidad (Loss Ratio): {loss_ratio:.1%}")
````

````python
# ── Filtrar filas (equivalente a WHERE en SQL) ──────────────────────────────

# Pólizas con siniestros
con_siniestros = df[df["siniestros_2025"] > 0]
print(f"Pólizas con siniestros: {len(con_siniestros)}")
print(con_siniestros[["nombre", "edad", "prima_anual",
                       "siniestros_2025", "monto_siniestros"]])
````

````python
# ── Crear columnas derivadas ─────────────────────────────────────────────────
# En actuaría frecuentemente creamos variables nuevas a partir de las existentes

df["cobertura_millon"] = df["suma_asegurada"] / 1_000_000
df["siniestralidad_individual"] = (
    df["monto_siniestros"] / df["prima_anual"]
).round(4)
df["grupo_edad"] = pd.cut(
    df["edad"],
    bins=[0, 30, 45, 60, 100],
    labels=["18-30", "31-45", "46-60", "61+"]
)

print("Columnas derivadas agregadas:")
df[["nombre", "edad", "grupo_edad", "cobertura_millon",
    "siniestralidad_individual"]].to_string(index=False)
````

## 7. Visualización Rápida

En las sesiones 11 y 12 profundizaremos en Matplotlib y Seaborn.

````python
# Configuración de estilo global (ejecutar una vez al inicio del notebook)
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({
    "figure.dpi": 120,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "grid.alpha": 0.3,
    "font.size": 10,
})
sns.set_palette("Set2")
print("✓ Estilo configurado")
````

````python
# ── Distribución de primas ───────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
fig.suptitle("Análisis de Cartera GMM — Distribución de Primas", 
             fontsize=13, fontweight="bold", y=1.02)

# Gráfica 1: Primas por asegurado
axes[0].bar(df["nombre"].str.split().str[0], df["prima_anual"],
            color="#0D7A8A", edgecolor="white", linewidth=0.5)
axes[0].set_title("Prima anual por asegurado")
axes[0].set_ylabel("Prima (MXN)")
axes[0].tick_params(axis='x', rotation=45)
for bar, val in zip(axes[0].patches, df["prima_anual"]):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 150,
                 f"${val:,.0f}", ha='center', va='bottom', fontsize=7)

# Gráfica 2: Prima vs Edad
axes[1].scatter(df["edad"], df["prima_anual"],
                s=df["suma_asegurada"]/5000, c="#F4A261",
                edgecolors="#1A5276", linewidths=0.8, alpha=0.85)
axes[1].set_title("Prima vs Edad(tamaño = suma asegurada)")
axes[1].set_xlabel("Edad")
axes[1].set_ylabel("Prima anual (MXN)")

# Gráfica 3: Distribución por grupo de edad
conteo = df["grupo_edad"].value_counts().sort_index()
axes[2].bar(conteo.index.astype(str), conteo.values,
            color=["#A8D8EA", "#0D7A8A", "#1A5276", "#0D1B3E"],
            edgecolor="white")
axes[2].set_title("Pólizas por grupo de edad")
axes[2].set_ylabel("Número de pólizas")
axes[2].set_xlabel("Grupo de edad")

plt.tight_layout()
plt.savefig("cartera_gmm_analisis.png", bbox_inches="tight", dpi=120)
plt.show()
print("✓ Gráfica guardada como 'cartera_gmm_analisis.png'")
````

````python
# ── Mapa de calor de correlaciones ───────────────────────────────────────────
# Muy usado en actuaría para explorar relaciones entre variables

cols_num = ["edad", "suma_asegurada", "prima_anual",
            "siniestros_2025", "monto_siniestros"]

fig, ax = plt.subplots(figsize=(7, 5))
correlacion = df[cols_num].corr()

sns.heatmap(
    correlacion, annot=True, fmt=".2f", cmap="coolwarm",
    center=0, square=True, linewidths=0.5,
    cbar_kws={"shrink": 0.8}, ax=ax
)
ax.set_title("Correlación entre variables — Cartera GMM",
             fontsize=12, fontweight="bold", pad=12)
plt.tight_layout()
plt.show()

print("Interpretación rápida:")
print(f"  Correlación prima ~ edad:        {correlacion.loc['prima_anual','edad']:.3f}")
print(f"  Correlación prima ~ suma aseg.:  {correlacion.loc['prima_anual','suma_asegurada']:.3f}")
print(f"  Correlación siniestros ~ edad:   {correlacion.loc['siniestros_2025','edad']:.3f}")
````
